```python
# Extracting integer value from 'Price' column
df['Price'] = df['Price'].str.extract('(\d+)').astype(int)

print(df)
```

```python
# Replacing 'new bus' with '0' in 'Bus Rating' column
df['Bus Rating'] = df['Bus Rating'].replace('new bus', '0')

# Converting 'Bus Rating' to numeric if needed
df['Bus Rating'] = pd.to_numeric(df['Bus Rating'], errors='coerce')

df.head()
```

```python
df.info()
```

```python
# Checking column types
print(df['Reaching Time'].dtype)
print(df['Departure Time'].dtype)

# Converting to timedelta
def convert_to_timedelta(time):
    if isinstance(time, str):  # If the time is in string format
        return pd.to_timedelta(time + ':00')
    elif isinstance(time, pd.Timestamp):  # If the time is in datetime format
        return pd.to_timedelta(f"{time.hour:02}:{time.minute:02}")
    else:
        raise ValueError("Unsupported time format")

df['Reaching Time'] = df['Reaching Time'].apply(convert_to_timedelta)
df['Departure Time'] = df['Departure Time'].apply(convert_to_timedelta)

# Verifying the result
print(df.head())
```

```python
df.info()
```

```python
df.head()
```

```python
def parse_duration(duration_str):
    try:
        # Cleaning the string, remove any unexpected characters
        duration_str = duration_str.replace('-', ' ').replace(',', ' ')
        parts = duration_str.split()
        
        # Initializing hours and minutes
        hours = 0
        minutes = 0

        # Extracting hours and minutes if present
        if 'h' in parts[0]:
            hours = int(parts[0].replace('h', ''))
        if 'm' in parts[1] if len(parts) > 1 else '':
            minutes = int(parts[1].replace('m', ''))
        
        # Creating timedelta
        return pd.Timedelta(hours=hours, minutes=minutes)
    
    except (IndexError, ValueError) as e:
        # Handling invalid formats by setting a default value or raise an error
        print(f"Error parsing '{duration_str}': {e}")
        return pd.NaT  # Using pd.NaT to denote missing values

# Applying the function to the 'Duration' column
df['Duration'] = df['Duration'].apply(parse_duration)

# Verifying the result
print(df.head())
print(df['Duration'].dtype)  
```

```python
import re

def standardize_ac_type(bus_type):
    # Defining regular expressions for each standard category
    ac_pattern = r"(?:\b|[^a-zA-Z0-9])(?:A/C|AC|A\.C|A/c|A\.C\.)(?:\b|[^a-zA-Z0-9])"
    non_ac_pattern = r"(?:\b|[^a-zA-Z0-9])(?:Non A/C|NON A/C|NON AC|NON A\.C|NON-AC)(?:\b|[^a-zA-Z0-9])"
    
    # Applying regular expressions to match the bus type
    if re.search(non_ac_pattern, bus_type, flags=re.IGNORECASE):
        return "Non A/C"
    elif re.search(ac_pattern, bus_type, flags=re.IGNORECASE):
        return "A/C"
    else:
        return "Other"  # For bus types that don't match any standard category ie,. buses that are not defined explicitly

# Applying the function to the 'Bus Type' column and create a new column
df['standardized_ac_type'] = df['Bus Type'].apply(standardize_ac_type)

print(df)
```

```python
# Getting unique bus types
unique_bus_types = df['Bus Type'].unique()

# Creating a new DataFrame with unique bus types and their standardized classification
classification_data = {'Bus Type': unique_bus_types}
classification_data['Standardized AC Type'] = [standardize_ac_type(bt) for bt in unique_bus_types]

df_classification = pd.DataFrame(classification_data)

# Saving the new DataFrame to a CSV file
df_classification.to_csv('bus_type_classification.csv', index=False)

print(df_classification)
```

```python
import re

def standardize_seat_type(bus_type):
    # Defining regular expressions for each standard category with non-alphanumeric boundaries
    sleeper_pattern = r"(?:\b|[^a-zA-Z0-9])(?:Sleeper|Executive|SLEEPER)(?:\b|[^a-zA-Z0-9])"
    seater_pattern = r"(?:\b|[^a-zA-Z0-9])(?:PUSH BACK|Seater|Semi Sleeper|Push Back|PushBack)(?:\b|[^a-zA-Z0-9])"
    seater_sleeper_pattern = seater_sleeper_pattern = r"(?:\b|[^a-zA-Z0-9])(?:Seater\s*/\s*Sleeper|Seater\s*/\s*Sleeper|SEATER\s*/\s*SLEEPER|SEATER\s*/\s*SLEEPER)(?:\b|[^a-zA-Z0-9])"
    
    # Applying regular expressions to match the bus type
    if re.search(seater_sleeper_pattern, bus_type, flags=re.IGNORECASE):
        return "Seater/Sleeper(2+1)"
    elif re.search(seater_pattern, bus_type, flags=re.IGNORECASE):
        return "Seater"
    elif re.search(sleeper_pattern, bus_type, flags=re.IGNORECASE):
        return "Sleeper"
    else:
        return "Other"  # For bus types that don't match any standard category ot buses that are not explicitly defined

# Applying the function to the 'Bus Type' column and create a new column
df['standardized_seat_type'] = df['Bus Type'].apply(standardize_seat_type)
df
```

```python
# Getting unique bus types
unique_bus_types = df['Bus Type'].unique()

# Creating a new DataFrame with unique bus types and their standardized classification
classification_data_seat = {'Bus Type': unique_bus_types}
classification_data_seat['Standardized Seat Type'] = [standardize_seat_type(bt) for bt in unique_bus_types]

df_classification_seat = pd.DataFrame(classification_data_seat)

# Saving the new DataFrame to a CSV file
df_classification_seat.to_csv('bus_seat_classification.csv', index=False)

print(df_classification_seat)
```
